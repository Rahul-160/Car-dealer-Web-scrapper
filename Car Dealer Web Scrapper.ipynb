Imports

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

HTTP Request

```python
website = 'https://www.cars.com/shopping/results/?stock_type=all&makes%5B%5D=mercedes_benz&models%5B%5D=&maximum_distance=30&zip='
```

Get Request

```python
response = requests.get(website)
```

Status Code

```python
response.status_code
```

Soup Object

```python
Soup = BeautifulSoup(response.content, 'html.parser')
```

```python
Soup
```

Results

```python
results =  Soup.find_all('div', {'class' : 'vehicle-card'})
```

```python
len(results)
```

```python
results[0]
```

Name

```python
results[0].find('h2').get_text()
```

```python
results[0].find('span', {'class' : 'sds-rating__count'}).get_text()
```

Rating Count

```python
results[0].find('span', {'class' : 'test1 sds-rating__link sds-button-link'}).get_text()
```

Price

```python
results[0].find('span', {'class' : 'primary-price'}).get_text()
```

```python
results[0].find('div', {'class' : 'dealer-name'}).get_text().strip()
```

Put everything together inside a For loop

```python
name = []
dealer_name = []
rating = []
rating_count = []
price = []

for result in results:
  try:
   name.append(result.find('h2').get_text())
  except:
   name.append('n/a')

  try:
   dealer_name.append(result.find('div', {'class' : 'dealer-name'}).get_text().strip())
  except:
   dealer_name.append('n/a')

  try:
   rating.append(result.find('span', {'class' : 'sds-rating__count'}).get_text())
  except:
   rating.append('n/a')

  try:
   rating_count.append(result.find('span', {'class' : 'test1 sds-rating__link sds-button-link'}).get_text())
  except:
   rating_count.append('n/a')

  try:
   price.append(result.find('span', {'class' : 'primary-price'}).get_text())
  except:
   price.append('n/a')
```

Create Pandas Dataframe

```python
car_dealer = pd.DataFrame({'Name' : name, 'Dealer Name' : dealer_name, 'Rating' : rating, 'Rating Count' : rating_count, 'Price' : price})
```

```python
car_dealer
```

Data Cleaning

```python
car_dealer['Rating Count'] = car_dealer['Rating Count'].apply(lambda x : x.strip('reviews)').strip('('))
```

```python
car_dealer
```

Output in Excel

```python
car_dealer.to_excel('single_page_car.xlsx', index = False)
```

Pagination


```python
name = []
dealer_name = []
rating = []
rating_count = []
price = []

for i in range(1, 21):
  website = 'https://www.cars.com/shopping/results/?dealer_id=&keyword=&list_price_max=&list_price_min=&makes[]=mercedes_benz&maximum_distance=30&monthly_payment=&page=' + str(i) + '&page_size=20&sort=best_match_desc&stock_type=all&year_max=&year_min=&zip='

  response = requests.get(website)

  Soup = BeautifulSoup(response.content, 'html.parser')

  results =  Soup.find_all('div', {'class' : 'vehicle-card'})
  for result in results:
   try:
    name.append(result.find('h2').get_text())
   except:
    name.append('n/a')

   try:
    dealer_name.append(result.find('div', {'class' : 'dealer-name'}).get_text().strip())
   except:
    dealer_name.append('n/a')

   try:
    rating.append(result.find('span', {'class' : 'sds-rating__count'}).get_text())
   except:
    rating.append('n/a')

   try:
    rating_count.append(result.find('span', {'class' : 'test1 sds-rating__link sds-button-link'}).get_text())
   except:
    rating_count.append('n/a')

   try:
    price.append(result.find('span', {'class' : 'primary-price'}).get_text())
   except:
    price.append('n/a')
```

```python
car_dealer = pd.DataFrame({'Name' : name, 'Dealer Name' : dealer_name, 'Rating' : rating, 'Rating Count' : rating_count, 'Price' : price})
```

```python
car_dealer['Rating Count'] = car_dealer['Rating Count'].apply(lambda x : x.strip('reviews)').strip('('))
```

```python
car_dealer
```

```python
car_dealer.to_excel('multiple_page_car.xlsx', index = False)
```
